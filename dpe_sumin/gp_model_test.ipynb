```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
import os
os.environ['CUDA_VISIBLE_DEVICES']='1'
```

```python
model = keras.models.load_model("gp_images/gp_model_epoch_80.h5")
```

```python
from image_loader import *

iml = Loader()

raw, clean = iml.load_from_npy_old()
```

```python
raw_test = raw[4503:4506]
clean_test = clean[4503:4506]
```

```python
yhat = model.predict(raw_test)
```

```python
plt.subplot(131)
plt.imshow(raw_test[0])
plt.subplot(132)
plt.imshow(tf.cast(yhat[0],np.uint8))
plt.subplot(133)
plt.imshow(cv2.cvtColor(clean_test[0],cv2.COLOR_BGR2RGB))
plt.show()
```
