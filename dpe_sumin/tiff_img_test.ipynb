```python
path = "/data/new/fivek_dataset/INPUT"
```

```python
import os
imglist = os.listdir(path)
imglist.sort()
```

```python
from matplotlib.image import imread
import matplotlib.pyplot as plt

rawlist = []
for list in imglist:
    rawlist.append(imread(path+"/%s"%list))
```

```python
plt.imshow(rawlist[4500])
plt.show()
```

```python
import numpy as np
import tensorflow as tf
from skimage.transform import resize

rsimg = tf.image.resize(rawlist[0],(512,512))
```

```python
plt.imshow(tf.cast(rsimg,tf.int32))
plt.show()
```

```python
new_raw = []
count = 0
for img in rawlist:
    new_raw.append(resize(img,(512,512)))
```

```python
np.save("new_raw",new_raw)
```

```python
new_clean = []

clean = []
path2 = "/data/new/fivek_dataset/CLEAR"
imglist = os.listdir(path2)
imglist.sort()

for list in imglist:
    clean.append(imread(path2+"/%s"%list))
for img in clean:
    new_clean.append(resize(img,(512,512)))
```

```python
np.save("new_clean",new_clean)
```
